```python
#DISPLAY THE SUM ON NULL VALUES IN EACH ROWS
import pandas as pd
data=pd.read_csv("SAMPLEIDS.csv")
df = pd.DataFrame(data)
null_sum_per_row = df.isnull().sum(axis=1)
print(null_sum_per_row)
```

```python
#DROP NULL VALUES
x=df.dropna(how='any')
x
```

```python
#FILL NULL VALUES WITH ffill or bfill METHOD
import pandas as pd
data = pd.read_csv("SAMPLEIDS.csv")
df = pd.DataFrame(data)
df_ffill = df.ffill()
df_bfill = df.bfill()
print("DataFrame after forward fill:")
print(df_ffill)
print("\nDataFrame after backward fill:")
print(df_bfill)
```

```python
#CALCULATE MEAN VALUE OF A COLUMN AND FILL IT WITH NULL 
import pandas as pd
data = pd.read_csv("SAMPLEIDS.csv")
df = pd.DataFrame(data)
mean_A = df['A'].mean()
df['A'] = df['A'].fillna(mean_A)
print("DataFrame after filling null values in column 'A' with mean value:")
print(df)
```

```python
#DROP NULL VALUES
x=df.dropna(how= 'any')
x
```
